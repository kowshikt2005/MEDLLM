```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    max_seq_length = SEQ_LEN,
    load_in_4bit = True,
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    lora_alpha = 16,

    target_modules = [
        "q_proj","k_proj","v_proj","o_proj",
        "gate_proj","up_proj","down_proj"
    ],

    lora_dropout = 0.1,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
)

trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
total     = sum(p.numel() for p in model.parameters())

print(f"Trainable: {trainable/1e6:.1f}M / {total/1e9:.1f}B  ({100*trainable/total:.2f}%)")
```

```python
from datasets import load_dataset
medmcqa     = load_dataset("openlifescienceai/medmcqa", split="train")
medmcqa_val = load_dataset("openlifescienceai/medmcqa", split="validation")  # for eval

print("Loading MedQA...")

medqa = load_dataset(
    "GBaker/MedQA-USMLE-4-options",
    split="train"
)
```

```python
medmcqa = medmcqa.filter(lambda x:
    x["cop"] in [0,1,2,3] and
    x["question"] and
    x["opa"] and x["opb"] and x["opc"] and x["opd"]
)

medmcqa = medmcqa.shuffle(seed=42).select(range(MEDMCQA_SAMPLE))
```

```python
OPTION_MAP = {0:"A",1:"B",2:"C",3:"D"}

PROMPT = """<s>[INST] You are an experienced physician and medical licensing exam tutor.

Carefully analyze the following multiple-choice question and determine the correct answer.

Question:
{question}

Options:
A) {A}
B) {B}
C) {C}
D) {D}

Think briefly about the medical concepts involved and choose the best answer.

Respond with:
Answer: <option letter>
Explanation: <short medical reasoning>

[/INST]
Answer: {answer}
Explanation: The correct answer is {answer}) {answer_text}.</s>
"""
```

```python
def format_medmcqa(x):

    letter = OPTION_MAP[int(x["cop"])]

    options = {
        "A": x["opa"],
        "B": x["opb"],
        "C": x["opc"],
        "D": x["opd"],
    }

    text = PROMPT.format(
        question=x["question"],
        A=x["opa"],
        B=x["opb"],
        C=x["opc"],
        D=x["opd"],
        answer=letter,
        answer_text=options[letter],
    )

    tokens = tokenizer(
        text,
        truncation=True,
        max_length=500,
    )

    return {"text": tokenizer.decode(tokens["input_ids"])}
```

```python
def format_medqa(x):

    letter = x["answer_idx"]
    opts = x["options"]

    text = PROMPT.format(
        question=x["question"],
        A=opts["A"],
        B=opts["B"],
        C=opts["C"],
        D=opts["D"],
        answer=letter,
        answer_text=opts[letter],
    )

    tokens = tokenizer(
        text,
        truncation=True,
        max_length=500,
    )

    return {"text": tokenizer.decode(tokens["input_ids"])}
```

```python
medmcqa_fmt = medmcqa.map(format_medmcqa, remove_columns=medmcqa.column_names)
medqa_fmt   = medqa.map(format_medqa,     remove_columns=medqa.column_names)
val_dataset = medmcqa_val.map(format_medmcqa, remove_columns=medmcqa_val.column_names)
val_dataset = val_dataset.select(range(500))  # 500 examples keeps eval fast (~3 min)

from datasets import concatenate_datasets
dataset = concatenate_datasets([medmcqa_fmt, medqa_fmt]).shuffle(seed=42)

print(f"Train: {len(dataset):,} examples")
print(f"Val:   {len(val_dataset):,} examples")
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported
```

```python
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    eval_dataset=val_dataset,
    dataset_text_field="text",
    max_seq_length=512,
    dataset_num_proc=2,
    packing=True,

    args=TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=16,
        num_train_epochs=1,
        max_steps=MAX_STEPS,
        warmup_steps=50,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=10,
        optim="adamw_8bit",
        weight_decay=0.05,
        lr_scheduler_type="cosine",
        output_dir="outputs",
        report_to="none",
        eval_strategy="steps",
        eval_steps=100,
        save_strategy="steps",
        save_steps=100,
        load_best_model_at_end=True,
        metric_for_best_model="eval_loss",
        greater_is_better=False,
    ),
)
```

```python
stats = trainer.train()

print("Training time:", stats.metrics["train_runtime"])
print("Loss:", stats.metrics["train_loss"])
```
